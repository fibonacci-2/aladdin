```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from datasets import load_dataset
import numpy as np
import os

ds = load_dataset("arbml/DART", split="train")

def analyze_dataset_stats(ds):
    print("=== DATASET STATISTICS ===")
    print(f"Total number of documents: {len(ds)}")
    
    # Extract text data
    texts = [doc["tweet_text"] for doc in ds]

    # Calculate basic statistics
    text_lengths = [len(text) for text in texts]
    word_counts = [len(text.split()) for text in texts]
    
    print(f"\nText Length Statistics (characters):")
    print(f"Mean: {np.mean(text_lengths):.2f}")
    print(f"Median: {np.median(text_lengths):.2f}")
    print(f"Min: {min(text_lengths)}")
    print(f"Max: {max(text_lengths)}")
    print(f"Std Dev: {np.std(text_lengths):.2f}")
    
    print(f"\nWord Count Statistics:")
    print(f"Mean: {np.mean(word_counts):.2f}")
    print(f"Median: {np.median(word_counts):.2f}")
    print(f"Min: {min(word_counts)}")
    print(f"Max: {max(word_counts)}")
    print(f"Std Dev: {np.std(word_counts):.2f}")
    
    # Sample texts
    print(f"\nSample texts:")
    for i in range(min(100, len(texts))):
        print(f"{i+1}. {texts[i][:100]}...")
    
    # Create visualizations
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Text length distribution
    axes[0, 0].hist(text_lengths, bins=50, alpha=0.7, color='skyblue')
    axes[0, 0].set_title('Distribution of Text Length (Characters)')
    axes[0, 0].set_xlabel('Character Count')
    axes[0, 0].set_ylabel('Frequency')
    
    # Word count distribution
    axes[0, 1].hist(word_counts, bins=50, alpha=0.7, color='lightgreen')
    axes[0, 1].set_title('Distribution of Word Count')
    axes[0, 1].set_xlabel('Word Count')
    axes[0, 1].set_ylabel('Frequency')
    
    # Box plot for text length
    axes[1, 0].boxplot(text_lengths)
    axes[1, 0].set_title('Text Length Box Plot')
    axes[1, 0].set_ylabel('Character Count')
    
    # Box plot for word count
    axes[1, 1].boxplot(word_counts)
    axes[1, 1].set_title('Word Count Box Plot')
    axes[1, 1].set_ylabel('Word Count')
    
    plt.tight_layout()
    plt.savefig(os.path.join(DATA_CACHE_DIR, 'dataset_stats.png'))
    plt.show()

# Add this after loading the dataset and before tokenization
print("Analyzing dataset...")
analyze_dataset_stats(ds)

# Continue with existing tokenization code...
all_tokens_np[token_count:token_count+len(tokens)] = tokens
token_count += len(tokens)
if progress_bar is None:
    progress_bar = tqdm(total=shard_size, unit="tokens", desc=f"Shard {shard_index}")
progress_bar.update(len(tokens))
```

```python
import arxiv
import requests

# Basic arXiv client
client = arxiv.Client()

# Search for recent papers
search = arxiv.Search(
  query="Deep Learning",
  max_results=5,
  sort_by=arxiv.SortCriterion.SubmittedDate
)
```
